In the dataset I selected, we get a glimpse into the citizenship status of participants. I specifically looked into the top 10 jurisdictions where female participation was the strongest. What stood out was that the majority of these participants were U.S. residents. This revelation highlights potential gender disparities related to citizenship, giving policymakers valuable information to address specific challenges faced by different gender groups within various citizenship categories.

These insights and visual representations offer a deep understanding of the dataset, empowering us to make decisions based on data and plan policies that are well-informed and targeted.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px



df = pd.read_csv("/content/Demographic_Statistics_By_Zip_Code.csv")
```

```python
#Selecting the dataset and sorting according to the highest number of female participants in the jurisdiction

df.set_index('JURISDICTION NAME', inplace= False)

df_res = df[['JURISDICTION NAME', 'COUNT FEMALE', 'COUNT MALE', 'COUNT PERMANENT RESIDENT ALIEN', 'COUNT US CITIZEN',
         'COUNT OTHER CITIZEN STATUS']]


df_res_sort = df_res.sort_values(by='COUNT FEMALE', ascending=False).head(10)
df_res_sort
```

```python
#Creating a horizontal bar graph to display Citizenship Status with the Highest Female Participation

color = ["blue", "magenta", "purple", "teal", "orange"]

df_res_sort.plot.barh(x='JURISDICTION NAME', y=['COUNT FEMALE' , 'COUNT MALE', 'COUNT PERMANENT RESIDENT ALIEN', 'COUNT US CITIZEN',
                                                'COUNT OTHER CITIZEN STATUS'],
             title='Top 10 Jurisdictions with the Highest Number of Female Participants', color = color)
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
```
